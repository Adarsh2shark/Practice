```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import ExtraTreesClassifier
```

```python
data = pd.read_csv("D://Non Acad data//data//global_heat_index.csv")
```

```python
data.head(10)
```

```python
clean1 = data[data['Hour']<6]     # for all the days where hour is less than 6 am is taken 
clean2 = data[data['Hour']>18]
```

```python
clean1.head(10)
```

```python
clean2.head(10)
```

```python
data = data.drop(clean1.index, axis =0)
data = data.drop(clean2.index, axis =0)
```

```python
data.info
```

```python
data.head(10) 
```

```python
# we want all column but solar radiation, those are our features
# 9 columns are needed and all rows needed
x = data.iloc[:, 0:10]
y = data.iloc[:, -1]   # -1 means last column
```

```python
from sklearn.preprocessing import MinMaxScaler
x = pd.DataFrame(MinMaxScaler().fit(x).transform(x))
```

```python
col  = ['Year', 'Month', 'Day','Hour','Dew Point','Temperature','Pressure','Relative Humidity','Wind Direction','Wind Speed' ]
```

```python
# EXTRA TREE CLASSIFIER
```

```python
model = ExtraTreesClassifier()
model.fit(x, y)
```

```python
imp = pd.Series(model.feature_importances_, index = col)
imp
```

```python
# VISUALISATION
```

```python
imp.nlargest(5).plot(kind = 'bar')
```
